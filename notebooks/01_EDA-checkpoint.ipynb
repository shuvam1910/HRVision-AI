```python
# ============================================
# Numerical and Categorical Features
# ============================================

numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_columns = df.select_dtypes(include=["object"]).columns.tolist()

print(f"Total Numerical Columns : {len(numerical_columns)}")
print(numerical_columns)

print("\n" + "="*70)

print(f"Total Categorical Columns : {len(categorical_columns)}")
print(categorical_columns)
```

### 📌 Observation

- The dataset contains both numerical and categorical features.
- Machine learning models require categorical variables to be converted into numerical form.

```python
# ============================================
# Unique Values in Categorical Columns
# ============================================

for col in categorical_columns:
    print("="*60)
    print(f"{col}")
    print("="*60)
    print(df[col].unique())
    print()
```

###  Observation

- Different categorical columns contain different categories.
- These categorical values must be encoded before training the model.

```python
# ============================================
# Remove EmployeeNumber
# ============================================

df.drop("EmployeeNumber", axis=1, inplace=True)

print("EmployeeNumber column removed successfully.")
```

###  Observation

- EmployeeNumber is only an employee identifier.
- It does not help in predicting employee attrition.

```python
# ============================================
# Encode Target Variable
# ============================================

df["Attrition"] = df["Attrition"].map({
    "Yes":1,
    "No":0
})

df["Attrition"].value_counts()
```

###  Observation

- The target variable has been converted into numerical values.
- Yes = 1
- No = 0

```python
# ============================================
# Label Encoding
# ============================================

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

for col in categorical_columns:

    if col != "Attrition":

        df[col] = label_encoder.fit_transform(df[col])

print("All categorical columns encoded successfully.")
```

###  Observation

- All categorical columns have been converted into numerical format using Label Encoding.

```python
df.head()
```

```python
df.info()
```

```python
# ============================================
# Feature Matrix and Target Variable
# ============================================

X = df.drop("Attrition", axis=1)

y = df["Attrition"]

print("Feature Shape :", X.shape)
print("Target Shape :", y.shape)
```

```python
# ============================================
# Train Test Split
# ============================================

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Shape :", X_train.shape)
print("Testing Shape :", X_test.shape)
```

###  Observation

- 80% of the data is used for training.
- 20% of the data is reserved for testing.

```python
# ============================================
# Feature Scaling
# ============================================

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)

print("Feature Scaling Completed Successfully.")
```

###  Observation

- Numerical features have been standardized using StandardScaler.
- Scaling helps algorithms such as KNN and SVM perform better.

```python
# ============================================
# Machine Learning Models
# ============================================

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)
```

```python
# Store model performance

model_results = {
    "Model": [],
    "Accuracy": [],
    "Precision": [],
    "Recall": [],
    "F1 Score": []
}
```

```python
# ============================================
# Logistic Regression
# ============================================

log_model = LogisticRegression(max_iter=1000)

log_model.fit(X_train_scaled, y_train)

y_pred = log_model.predict(X_test_scaled)

acc = accuracy_score(y_test, y_pred)
pre = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

model_results["Model"].append("Logistic Regression")
model_results["Accuracy"].append(acc)
model_results["Precision"].append(pre)
model_results["Recall"].append(rec)
model_results["F1 Score"].append(f1)

print(classification_report(y_test, y_pred))
```

###  Observation

- Logistic Regression provides a strong baseline model.
- Performance will be compared with other algorithms.

```python
plt.figure(figsize=(5,4))

sns.heatmap(
    confusion_matrix(y_test, y_pred),
    annot=True,
    fmt="d",
    cmap="Blues"
)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Logistic Regression Confusion Matrix")

plt.show()
```

###  Observation

- The confusion matrix shows the number of correct and incorrect predictions.
- It helps evaluate the model beyond overall accuracy.

```python
# ============================================
# Decision Tree
# ============================================

dt_model = DecisionTreeClassifier(random_state=42)

dt_model.fit(X_train, y_train)

y_pred = dt_model.predict(X_test)

acc = accuracy_score(y_test, y_pred)
pre = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

model_results["Model"].append("Decision Tree")
model_results["Accuracy"].append(acc)
model_results["Precision"].append(pre)
model_results["Recall"].append(rec)
model_results["F1 Score"].append(f1)

print(classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(5,4))

sns.heatmap(
    confusion_matrix(y_test, y_pred),
    annot=True,
    fmt="d",
    cmap="Greens"
)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Decision Tree Confusion Matrix")

plt.show()
```

```python
# ============================================
# Random Forest
# ============================================

rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42
)

rf_model.fit(X_train, y_train)

y_pred = rf_model.predict(X_test)

acc = accuracy_score(y_test, y_pred)
pre = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

model_results["Model"].append("Random Forest")
model_results["Accuracy"].append(acc)
model_results["Precision"].append(pre)
model_results["Recall"].append(rec)
model_results["F1 Score"].append(f1)

print(classification_report(y_test,y_pred))
```

###  Observation

- Random Forest combines multiple decision trees.
- It generally provides better accuracy and reduces overfitting.

```python
plt.figure(figsize=(5,4))

sns.heatmap(
    confusion_matrix(y_test,y_pred),
    annot=True,
    fmt="d",
    cmap="Oranges"
)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Random Forest Confusion Matrix")

plt.show()
```

```python
# ============================================
# K Nearest Neighbors
# ============================================

knn_model = KNeighborsClassifier(n_neighbors=5)

knn_model.fit(X_train_scaled,y_train)

y_pred = knn_model.predict(X_test_scaled)

acc = accuracy_score(y_test,y_pred)
pre = precision_score(y_test,y_pred)
rec = recall_score(y_test,y_pred)
f1 = f1_score(y_test,y_pred)

model_results["Model"].append("KNN")
model_results["Accuracy"].append(acc)
model_results["Precision"].append(pre)
model_results["Recall"].append(rec)
model_results["F1 Score"].append(f1)

print(classification_report(y_test,y_pred))
```

###  Observation

- KNN predicts based on the nearest neighboring data points.
- Feature scaling is important for KNN performance.

```python
plt.figure(figsize=(5,4))

sns.heatmap(
    confusion_matrix(y_test,y_pred),
    annot=True,
    fmt="d",
    cmap="Purples"
)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("KNN Confusion Matrix")

plt.show()
```

```python
# ============================================
# Support Vector Machine
# ============================================

svm_model = SVC()

svm_model.fit(X_train_scaled,y_train)

y_pred = svm_model.predict(X_test_scaled)

acc = accuracy_score(y_test,y_pred)
pre = precision_score(y_test,y_pred)
rec = recall_score(y_test,y_pred)
f1 = f1_score(y_test,y_pred)

model_results["Model"].append("SVM")
model_results["Accuracy"].append(acc)
model_results["Precision"].append(pre)
model_results["Recall"].append(rec)
model_results["F1 Score"].append(f1)

print(classification_report(y_test,y_pred))
```

###  Observation

- SVM attempts to find the optimal decision boundary.
- It often performs well for binary classification problems.

```python
plt.figure(figsize=(5,4))

sns.heatmap(
    confusion_matrix(y_test,y_pred),
    annot=True,
    fmt="d",
    cmap="Reds"
)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("SVM Confusion Matrix")

plt.show()
```
